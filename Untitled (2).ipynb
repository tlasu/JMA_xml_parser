```python

def send(WEBHOOK_URL,data={"text": "AWS からのアクセスHello World_request moduleによるテスト"}):
    r=rq.post(WEBHOOK_URL,json=data,
              headers={'Content-Type': 'application/json'})
    print(r.status_code)

def lambda_handler(event, context):
    # TODO implement
    WEBHOOK_URL = os.environ['WEBHOOK_URL']
    #print(WEBHOOK_URL)
    print('========start========')
    old_data=get_ids_old_data()
    url="http://www.data.jma.go.jp/developer/xml/feed/{}.xml".format(target_file)
    print(url)
    time_c=dt.datetime.now().strftime("%Y%m%d%H%M")
    r=rq.get(url)
    data=r.content
    
    f1="latest_{}.xml".format(target_file)
    f2="{}_{}.xml".format(time_c,target_file)
    
    with open("/tmp/"+f1,"wb") as f:
        f.write(data)
    with open("/tmp/"+f2,"wb") as f:
        f.write(data)
    s3.meta.client.upload_file('/tmp/'+f1, bucket, f1)
    s3.meta.client.upload_file('/tmp/'+f2, bucket, 'xml/'+f2)
    
    f = io.BytesIO(data)
    tree = ET.parse(f)
    root = tree.getroot()
    entrys=root.findall("{http://www.w3.org/2005/Atom}entry")
    target_ent=[]
    for ent in entrys:
        if (ent.find('{http://www.w3.org/2005/Atom}title').text=="指定河川洪水予報"):
            target_ent.append(ent)

    of1="latest_flood_{}.json".format(target_file)
    of2="latest_flood_{}.csv".format(target_file)
    of3=time_c+"flood_{}.json".format(target_file)
    of4=time_c+"flood_{}.csv".format(target_file)
    
    if(len(target_ent)!=0):
        ent_class=[bousai_Xml(a,old_data=old_data) for a in target_ent]
        out_data=pd.DataFrame([vars(ent_c) for ent_c in ent_class])
        
        #ファイルの出力
        out_data.to_json('/tmp/'+of1)
        out_data.to_csv('/tmp/'+of2)
        out_data.to_json('/tmp/'+of3)
        out_data.to_csv('/tmp/'+of4)        
        for _f in [of1,of2]:
            s3.meta.client.upload_file('/tmp/'+_f, bucket, _f)
        
        s3.meta.client.upload_file('/tmp/'+of3, bucket, 'json/'+of3)    
        s3.meta.client.upload_file('/tmp/'+of4, bucket, 'csv/'+of4)    
        new_data_number=0
        for ent_c in ent_class[::-1]:
            if ent_c.new:
                new_data_number+=1
                ent_c.send(WEBHOOK_URL)
        info="Published {} information and New {} information".format(len(target_ent),new_data_number)
    else:
        info="no information"
    
    #tmp配下を削除
    call('rm -rf /tmp/*', shell=True) 
    print(info)
    return {
        'statusCode': 200,
        'body': json.dumps(info)
    }
```

```python
def get_ids_old_data():
    """#前回のデータからid一覧のみを取得"""
    try:
        if is_aws_lambda:
            src_obj = s3.Object(bucket,'latest_extra.xml')
            body_in = src_obj.get()['Body'].read().decode('utf8')
            f = io.StringIO(body_in)
        else:
            f = target_file
            
        tree = ET.parse(f)
        root = tree.getroot()
        entrys=root.findall("{http://www.w3.org/2005/Atom}entry")
        target_ent=[]
        for ent in entrys:
            if (ent.find('{http://www.w3.org/2005/Atom}title').text=="指定河川洪水予報"):
                target_ent.append(ent.find('{http://www.w3.org/2005/Atom}id').text)
        return pd.DataFrame(target_ent,columns=["id"])
    except Exception as e:
        print("get_ids_old_data_error", e.args)
        return pd.DataFrame()
```
